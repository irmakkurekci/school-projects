```python
import numpy as np                 
import matplotlib.pyplot as plt    

#This is the empty image array on which the lines will be drawn.
img=np.ones((600,600,3))

#Function DrawLine() is created to calculate and create line pixels.
def DrawLine(x1,y1,x2,y2):
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    if x1<x2:
        xs=1 
    else:
        xs=-1   
    if y1<y2:
        ys=1
    else:
        ys=-1
    # The variables are created to store pixel coordinates. 
    x=x1
    y=y1
    
#To Draw Horizontal Line
  
    if dx>dy:
        p=2*dy-dx 
        
        while x!=x2:
            x=x+xs
            
            if p > 0:
                y=y+ys
                p=p+2*dy-2*dx
                
            else:
                p=p+2*dy  
                img[y,x]= 0

#To Draw Vertical Line                
    if dx<dy:
        p=2*dx-dy
        
        while y!=y2:
            y=y+ys
            
            if p > 0:
                x=x+xs
                p=p+2*dx-2*dy
                
            else:
                p=p+2*dx
                img[y,x]= 0
     

    return;

#if final result is not clear then please uncomment the code below to view it properly in new window.
# NOTE! please restart the kernel to unable it.

#%matplotlib qt

#These parameters can change the width, height as well as number of pixels per inch.
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams["figure.dpi"] = 80

#This will create number 1

DrawLine(100,200,100,400)
DrawLine(50,280,100,200)
DrawLine(50,400,150,400)


#This will create number 9

DrawLine(200,200,300,200)
DrawLine(300,400,300,200)
DrawLine(200,200,200,300)
DrawLine(200,300,300,300)
DrawLine(200,400,300,400)

#This will create number 4

DrawLine(400,200,400,300)
DrawLine(400,300,500,300)
DrawLine(500,300,500,200)
DrawLine(500,300,500,400)









plt.imshow(img)
plt.show()
```
