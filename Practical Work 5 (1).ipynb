```python
import numpy as np
from numpy import asarray

import cv2

import matplotlib.pyplot as plt

image_A = cv2.imread('black wolf.jpg')
image_B = cv2.imread('white wolf.jpg')

im_rgb_A = cv2.cvtColor(image_A, cv2.COLOR_BGR2RGB)
im_rgb_B = cv2.cvtColor(image_B, cv2.COLOR_BGR2RGB)

resized_img_A = cv2.resize(im_rgb_A, (500,500))
resized_img_B = cv2.resize(im_rgb_B, (500,500))

image_norm_A = resized_img_A/255.0
image_norm_B = resized_img_B/255.0


linearburn = (image_norm_A + (image_norm_B - 1))

colorburn = 1 - ((1-image_norm_B)/image_norm_A)


fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(resized_img_A)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(resized_img_B)

fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(3,3,1)
ax1.title.set_text('input 1')
ax1.imshow(image_norm_A)
ax1.axis('off')

ax2 = fig.add_subplot(3,3,2)
ax2.title.set_text('input 2')
ax2.imshow(image_norm_B)
ax2.axis('off')

ax3 = fig.add_subplot(3,3,3)
ax3.title.set_text('Subtract/Linear Burn')
ax3.imshow(linearburn)
ax3.axis('off')

ax4 = fig.add_subplot(3,3,4)
ax4.title.set_text('input 1')
ax4.imshow(image_norm_A)
ax4.axis('off')

ax2 = fig.add_subplot(3,3,5)
ax2.title.set_text('input 2')
ax2.imshow(image_norm_B)
ax2.axis('off')

ax5 = fig.add_subplot(3,3,6)
ax5.title.set_text('Burn/Color Burn')
ax5.imshow(colorburn)
ax5.axis('off')

```
