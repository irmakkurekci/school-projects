```python
import numpy as np
import matplotlib.pyplot as plt
import math

img = np.ones((2000,2000,3))

point1 = (600,400)
point2 = (400,400)
point3 = (400,600)
point4 = (600,600)

def MoveObject(move_x, move_y):
    x1 = point1[0] + move_x
    y1 = point1[1] + move_y
    x2 = point2[0] + move_x
    y2 = point2[1] + move_y
    x3 = point3[0] + move_x
    y3 = point3[1] + move_y
    x4 = point4[0] + move_x
    y4 = point4[1] + move_y
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    

def ScaleObject(scaleX,scaleY):
    x1 = point1[0] * scaleX
    y1 = point1[1] * scaleY
    x2 = point2[0] * scaleX
    y2 = point2[1] * scaleY
    x3 = point3[0] * scaleX
    y3 = point3[1] * scaleY
    x4 = point4[0] * scaleX
    y4 = point4[1] * scaleY
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
    
def RotateObject(a):
    x1 = point1[0] * (math.cos(a)) - point1[1] * (math.sin(a))
    y1 = point1[0] * (math.sin(a)) + point1[1] * (math.cos(a))
    x2 = point2[0] * (math.cos(a)) - point2[1] * (math.sin(a))
    y2 = point2[0] * (math.sin(a)) + point2[1] * (math.cos(a))
    x3 = point3[0] * (math.cos(a)) - point3[1] * (math.sin(a))
    y3 = point3[0] * (math.sin(a)) + point3[1] * (math.cos(a))
    x4 = point4[0] * (math.cos(a)) - point4[1] * (math.sin(a))
    y4 = point4[0] * (math.sin(a)) + point4[1] * (math.cos(a))
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
plt.figure(figsize=(10,8), dpi = 100, facecolor='w'),
plt.imshow(img)

MoveObject(300,100)
ScaleObject(3,2)
RotateObject(math.radians(15))
    
plt.show()
   

    
    
```
