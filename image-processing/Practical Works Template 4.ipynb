Based on the last 3 digits of your student ID number, re-colour any image that has been resized to width and height of 500 (500,500) with every N-th pixel on the odd row and every M-th pixel on the even row changed to a certain colour (green, red, black etc.)  

The N an M value and the colour the pixels need to be changed to can be found in the variant list available in the course description in ORTUS.
   
Example: N = 30, M = 10, color = Red

My variant is 194, and i will do N = 10, M = 30, color = White


```python
from PIL import Image
import matplotlib.pyplot as plt
image = Image.open('quantumphysics.jpg')
resized_img = image.resize((500,500))
pixels = resized_img.load()

for x in range(0,500):
    for y in range(0,500):
        
        if x%10 == 0 and y%2!=0: 
            pixels[x,y] = (255, 255, 255) #white
            
        if x%30 == 0 and y%2==0:
            pixels[x,y] = (255, 255, 255) #white


%matplotlib notebook
plt.figure(figsize=(7,7))
plt.imshow(resized_img)
```
