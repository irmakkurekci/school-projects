## TASK OF WORK

Draw a single digit with one or more Bezier curves (based on the variant). The list of variants is available in ORTUS under the 4th practical lesson!

The variant number corresponds to the last 3 digits of the student ID number. Possible variations:

number (3, 6, 8, 9),
amount of points (4-8 control points).


For example, if the last 3 digits of your student ID number are 154, then you must select variant 154 from the list of variant. The first column in the list of variants shows the number that needs to be drawn with Bezier curves and the second column shows how many control points the drawn Bezier curves should consist of.

Variant - 194
Number - 8
Control Points - 6


```python
import numpy as np               
import matplotlib.pyplot as plt  

img = np.ones((600,600,3))

def DrawBezier(x0,y0, x1,y1, x2,y2, x3,y3, x4,y4, x5,y5):
    
    x = x0
    y = y0

    ts = 0.001
    
    t = 0
    while t <= 1.0:
        
        B0 = 1 * np.power(1-t,5-0) * np.power(t,0)
        B1 = 5 * np.power(1-t,5-1) * np.power(t,1)
        B2 = 10 * np.power(1-t,5-2) * np.power(t,2)
        B3 = 10 * np.power(1-t,5-3) * np.power(t,3)
        B4 = 5 * np.power(1-t,5-4) * np.power(t,4)
        B5 = 1 * np.power(1-t,5-5) * np.power(t,5)


        x = int(x0*B0 + x1*B1 + x2*B2 + x3*B3 + x4*B4 + x5*B5)
        y = int(y0*B0 + y1*B1 + y2*B2 + y3*B3 + y4*B4 + y5*B5)
        
        img[x,y] = (1,0,0)
        
        t=t+ts 
        
    return
   
DrawBezier(200,400, 200,200, 400,400, 200,200, 300,200, 200,400)

    
%matplotlib notebook
plt.figure(figsize=(10,8), dpi = 100, facecolor='w')
plt.imshow(img)
plt.show()
```
